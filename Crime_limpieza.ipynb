```python
from sklearn.impute import KNNImputer

# Funci칩n para imputar latitud y longitud usando KNN con KNeighborsRegressor
def impute_lat_lon(df):
    # Concatenar latitud y longitud en una sola columna
    df['coordenada'] = df['latitud'].astype(str) + ',' + df['longitud'].astype(str)

    # Crear instancia del imputador KNN
    imputer = KNNImputer(n_neighbors=5)

    # Imputar latitud y longitud
    df[['latitud', 'longitud']] = imputer.fit_transform(df[['latitud', 'longitud']])

    # Eliminar la columna 'coordenada' ya que no es necesaria despu칠s de la imputaci칩n
    df.drop(columns=['coordenada'], inplace=True)

    return df

# Aplicar la funci칩n para imputar latitud y longitud
df = impute_lat_lon(df)
```

```python
# Reviso si se llenaron valores faltantes en latitud y longitud.

print("Resumen de valores faltantes luego de llenado faltantes:", df.isnull().sum())
```

# Genero intervalos de latitud-longitud para asignar barrios.

```python
from collections import defaultdict

# Definir una funci칩n para generar rangos de latitud y longitud
def generate_ranges(lat, lon, precision):
    lat_range = round(float(lat) / precision) * precision
    lon_range = round(float(lon) / precision) * precision
    return (lat_range, lon_range)

# Definir la precisi칩n deseada para los rangos
precision = 0.01  # Ajusta este valor seg칰n la precisi칩n requerida

# Creamos un diccionario donde cada barrio est치 asociado a un rango de coordenadas (latitud, longitud)
barrio_to_ranges = defaultdict(list)
for index, row in df.dropna(subset=['latitud', 'longitud', 'barrio']).iterrows():
    lat_range, lon_range = generate_ranges(row['latitud'], row['longitud'], precision)
    barrio_to_ranges[row['barrio']].append((lat_range, lon_range))

# Asignar barrio basado en los rangos de coordenadas asociados a cada barrio
for i, row in df.iterrows():
    if pd.isnull(row['barrio']):
        lat_range, lon_range = generate_ranges(row['latitud'], row['longitud'], precision)
        for barrio, ranges in barrio_to_ranges.items():
            if (lat_range, lon_range) in ranges:
                df.at[i, 'barrio'] = barrio
                break

print(df)

```
